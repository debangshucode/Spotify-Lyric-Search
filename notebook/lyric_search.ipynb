```python
import pandas as pd

from src.preprocess import preprocess_text
from src.vectorizer import create_vectorizer
from src.search import predict_song
```

```python
df = pd.read_csv("../data/spotify_lyrics.csv")

df = df[['artist', 'song', 'text']]
df.dropna(inplace=True)
```

```python
df['clean_lyrics'] = df['text'].apply(preprocess_text)
```

```python
vectorizer, lyric_vectors = create_vectorizer(df['clean_lyrics'])
```

```python
input_snippet = "Counting my pride"

result = predict_song(
    input_snippet,
    vectorizer,
    lyric_vectors,
    df
)

result
```

```python
correct = 0
total = 100

for i in range(total):
    row = df.iloc[i]

    prediction = predict_song(
        row['text'][:200],  
        vectorizer,
        lyric_vectors,
        df
    )

    if prediction['song'] == row['song']:
        correct += 1

accuracy = correct / total
print("Accuracy:", accuracy)
```
